Analysing Longitude and Latitude

```python


# Convert latitude and longitude columns to numeric, if necessary
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=df)
plt.title('Geographical Distribution of Data Points')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

The scatter plot shows a dense cluster of data points around lower latitude and longitude values, with a few outliers spread across higher longitude values. This indicates most data points are geographically concentrated in a specific area

```python


# Assuming df is your dataframe with 'latitude' and 'longitude' columns

# Group by latitude and longitude and count the occurrences
crime_locations = df.groupby(['Latitude', 'Longitude']).size().reset_index(name='count')

# Sort the values to find the locations with the most crimes
crime_locations = crime_locations.sort_values(by='count', ascending=False)

# Display the top locations
top_crime_locations = crime_locations.head(10)
print(top_crime_locations)
```

1.The most frequent latitude and longitude values (0.000000, 0.000000) likely represent missing or placeholder data, as they occur 1,181,061 times.
2.The next most frequent location is (6.752783, 68.177512) with 802 occurrences, indicating a high concentration of crimes at this location.
